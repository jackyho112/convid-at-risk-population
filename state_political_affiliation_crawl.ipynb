```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import ast
```

```python
main_request = requests.get("https://en.wikipedia.org/wiki/Political_party_strength_in_U.S._states")
main_page = BeautifulSoup(main_request.text)
```

```python
states = main_page.select('.wikitable.sortable tr')[1:]
```

```python
state_names = []
house_blue_perc = []
```

```python
for state in states:
    rows = state.select('td')
    state_name = rows[0].text.replace("\n", "")
    
    if state_name == 'Nebraska':
        continue
    
    state_names.append(state_name)
    
    party_in_power, party_ratio = rows[3].text.replace("\n", "").split(' ')
    
    party_numbers = list(map(lambda x: int(''.join(list(filter(str.isdigit, x)))), party_ratio.split('â€“')))
    
    if party_in_power == 'Republican':
        house_blue_perc.append(round(party_numbers[1]/sum(party_numbers)*100, 2))
    else:
        house_blue_perc.append(round(party_numbers[0]/sum(party_numbers)*100, 2))
```

```python
df = pd.DataFrame({
    'state': state_names,
    'state_house_blue_perc': house_blue_perc
})
```
